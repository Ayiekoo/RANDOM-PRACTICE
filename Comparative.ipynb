```python
import matplotlib.pyplot as plt

# Data
categories = ['Combined Therapy', 'CBT', 'Sertraline']
values = [53.6, 39.3, 21.4]
colors = ['tan', 'aqua', 'darkseagreen']

# Create the pie chart
plt.figure(figsize=(8, 8))
plt.pie(values, labels=categories, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Therapy Effectiveness')
plt.show()
```
