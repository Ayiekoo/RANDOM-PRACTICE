```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Setting seed for reproducibility
np.random.seed(42)

# Parameters
num_patients = 10
num_weeks = 26

# Generate random starting Y-BOCS scores between 25 and 40
start_scores = np.random.randint(25, 41, size=num_patients)

# Generate random weekly improvements between 0 and 2 points per week
weekly_improvements = np.random.uniform(0.5, 1.5, (num_patients, num_weeks))

# Initialize data frame to store scores
data = {
    'Week': np.arange(1, num_weeks + 1)
}
for patient in range(num_patients):
    scores = start_scores[patient] - np.cumsum(weekly_improvements[patient])
    scores[scores < 0] = 0  # Y-BOCS scores cannot be negative
    data[f'Patient {patient + 1}'] = scores

# Convert to DataFrame
df = pd.DataFrame(data)

# Plotting the data
plt.figure(figsize=(14, 8))
for patient in range(num_patients):
    plt.plot(df['Week'], df[f'Patient {patient + 1}'], label=f'Patient {patient + 1}')

# Adding labels and title
plt.xlabel('Week')
plt.ylabel('Y-BOCS Score')
plt.title('Weekly Y-BOCS Scores Over 26 Weeks of Treatment')
plt.legend(loc='upper right')
plt.grid(True)

# Showing the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data for comorbid conditions with OCD
conditions = ['Anxiety Disorder', 'Mood Disorder', 'ADHD', 'Substance Use Disorder']
percentages = [76, 63, 56, 39]

# Plotting the data
plt.figure(figsize=(10, 6))
bars = plt.barh(conditions, percentages, color=['tan', 'aqua', 'darkseagreen', 'cyan'])

# Adding data labels
for bar in bars:
    plt.text(bar.get_width() + 2, bar.get_y() + bar.get_height()/2, f'{bar.get_width()}%',
             ha='center', va='center', color='black', fontsize=12)

# Adding labels and title
plt.xlabel('Percentage of People with OCD (%)')
plt.title('Prevalence of Comorbid Conditions in People with OCD')
plt.xlim(0, 100)
plt.grid(True, axis='x', linestyle='--', alpha=0.7)

# Showing the plot
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Define categories and their probabilities
categories = ['Absent', 'Mild', 'Moderate', 'Severe', 'Extreme']
probabilities = [0.05, 0.25, 0.30, 0.25, 0.15]

# Define genders and their probabilities
genders = ['Male', 'Female']
gender_probabilities = [0.45, 0.55]

# Define ethnicities and their probabilities
ethnicities = ['Caucasian', 'African American', 'Hispanic', 'Asian', 'Other']
ethnicity_probabilities = [0.50, 0.15, 0.20, 0.10, 0.05]

# Number of samples
num_samples = 100

# Generate random data
np.random.seed(42)
data = {
    'Patient_ID': range(1, num_samples + 1),
    'YBOCS_Category': np.random.choice(categories, num_samples, p=probabilities),
    'Gender': np.random.choice(genders, num_samples, p=gender_probabilities),
    'Ethnicity': np.random.choice(ethnicities, num_samples, p=ethnicity_probabilities)
}

# Convert to DataFrame
df = pd.DataFrame(data)

# Display the first few rows of the dataframe
print(df.head())

# Create summary tables
summary_gender = df.groupby(['YBOCS_Category', 'Gender']).size().unstack(fill_value=0)
summary_ethnicity = df.groupby(['YBOCS_Category', 'Ethnicity']).size().unstack(fill_value=0)

# Define the colors
colors = ['tan', 'aqua', 'cyan', 'darkseagreen']

# Plotting the data for Gender
plt.figure(figsize=(12, 6))
summary_gender.plot(kind='bar', stacked=True, color=colors[:2])
plt.title('YBOCS Category Distribution by Gender')
plt.xlabel('YBOCS Category')
plt.ylabel('Number of Patients')
plt.legend(title='Gender')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Plotting the data for Ethnicity
plt.figure(figsize=(12, 6))
summary_ethnicity.plot(kind='bar', stacked=True, color=colors)
plt.title('YBOCS Category Distribution by Ethnicity')
plt.xlabel('YBOCS Category')
plt.ylabel('Number of Patients')
plt.legend(title='Ethnicity')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
